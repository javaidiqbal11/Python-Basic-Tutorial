#### Updating a dictionary
##### Just like list, dictionaries are mutable


```python
# Just like list, dictionaries are mutable
alphabet = {'a':'apple', 'b':'ball', 'c':'cat'  }

alphabet.update( {'b': 'bat'  }  )
print(alphabet)

alphabet.update( {'d': 'dog'  }  )
print(alphabet)
```

```python
# more like list syntax
alphabet = {'a':'apple', 'b':'ball', 'c':'cat'  }
alphabet['c'] = 'car'
print(alphabet)

alphabet['d'] = 'dog'
print(alphabet)
```

### Shallow/Reference copy vs. Deep copy

```python
# Reference copy
x = {'a':'apple', 'b':'ball', 'c':'cat'  }
y = x # both refers to the same dictionary object 

y.update( {'b':'bat'} )
x.update( {'d':'dog'} ) 
print(y)
print(x)

```

```python
#list
x = ['apple', 'bat', 'cat']
y = x # both refers to the same list object
y[2]='car'
print(y)
print(x)
```

```python
# Deep copy
#How can you have a copy of object that is not associated with each other
#list
x = ['apple', 'bat', 'cat']
y = x[:] #x[start:end:step] # deep copy (both x and y refers to two different lists)
y[2] = 'car'
print(y)
print(x)
```

```python
# creating a deep copy for dictionary
import copy
x = {'a':'apple', 'b':'ball', 'c':'cat'  }
y = copy.deepcopy(x)
y.update( {'d':'dog'})
print(x)
print(y)
```

```python
# creating a deep copy for dictionary
x = {'a':'apple', 'b':'ball', 'c':'cat' }
y = {}
for i in x:
    print(i, x[i])
    y.update( {i:x[i]} )

y.update({'d':'dog'})
print(y)
print(x)
```

### Week 1 Topic: Iterations

#### while loop

```python
# 1. start
# 2. end, condition
# 3. increment or decrement 
#     (so that you can get closer to the end in every iteration/step)

# infinite loop because we do not add increment
n = 0 # start
while (n<5): # condition n<5
    print(n)


```

```python
n = 0 # start
while (n<5): # condition n<5
    print(n)
    n=n+1  # increment
```

```python
n = 10 # start
while (n>=1): # condition stop at 1 (n>0)
    print(n)
    n=n-1  # decrement
```
